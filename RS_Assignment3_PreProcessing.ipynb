## Clean the text of description

```python
import nltk
nltk.download('omw-1.4')
nltk.download("stopwords")
nltk.download("punkt")
nltk.download("wordnet")
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

def clean_text(text):
    """
    Clean the text by removing punctuation, stopwords, and converting to lowercase.
    """
    # Remove punctuation
    text = re.sub(r"[^\w\s]", "", text)
    
    # Convert to lowercase
    text = text.lower()

    # Remove non-English characters
    text = re.sub(r"[^\x00-\x7F]+", "", text)
    
    # Tokenize the text
    words = word_tokenize(text)
    
    # Remove stopwords
    stop_words = set(stopwords.words("english"))
    words = [word for word in words if word not in stop_words]
    
    # Perform lemmatization
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]
    
    # Join the words back into a single string
    cleaned_text = " ".join(words)
    
    return cleaned_text
```

```python
df1["Synopsis"] = df["description"].apply(clean_text)
```

# Creating Word2Vec for synopsis

```python
df = pd.read_csv("/content/drive/MyDrive/RS Data/Assignment 3/Processed_data/Books_final_dataset.csv")
df
```

```python
df.isna().sum()
```

```python
import nltk

nltk.download("punkt")
nltk.download("wordnet")
```

```python
import numpy as np
```

```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize

# Tokenize the sentences
tokenized_sentences = [word_tokenize(sentence) for sentence in df["Synopsis"]]

# Train the Word2Vec model
model = Word2Vec(tokenized_sentences, vector_size=100, window=5, min_count=1)
```

```python
book_vectors = []
for i in range(len(df["Synopsis"])):
  words = df["Synopsis"][i].split()

  # Initialize the document representation vector
  document_vector = np.zeros((1,model.vector_size))

  # For each word in the document, get its word embedding and sum it with the document representation vector
  for word in words:
      if word in model.wv.key_to_index:
          document_vector += model.wv[word]

  # Divide the document representation vector by the number of words in the document to get the average word embedding
  document_vector = document_vector / len(words)
  book_vectors.append(document_vector)
```

```python
df4 = pd.DataFrame({"Word_vec":book_vectors})
```

```python
df4
```

```python
df['Word_Vec'] = df4
```

```python
import numpy as np
import pandas as pd

arr = np.vstack(df['Word_Vec'].values)

# Reshape the array to the desired shape
arr = arr.reshape(-1, 100)
arr.shape
```

```python
df_word2vec = pd.DataFrame(arr, columns=['feature_{}'.format(i+1) for i in range(100)])
df5 = pd.concat([df, df_word2vec], axis=1)

# drop the original 'word2vec' column
df5 = df5.drop('Word_Vec', axis=1)
```
