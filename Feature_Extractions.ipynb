```python
import pandas as pd
import time
from googleapiclient.discovery import build

# Define function to get book info from Google Books API
def get_book_info(title):
    service = build('books', 'v1', developerKey='AIzaSyAzrSbklmN2jXfwoylPcP3gDQJOy_B2ZYc')
    result = service.volumes().list(q=title).execute()
    if 'items' in result:
        book = result['items'][0]['volumeInfo']
        language = book.get('language', '')
        page_count = book.get('pageCount', '')
        categories = book.get('categories', [])
        description = book.get('description', '')
        return {'language': language, 'page_count': page_count, 'categories': categories, 'description': description}
    return None


# Define new columns for genre and description
books_details['language'] = ''
books_details['page_count'] = ''
books_details['categories'] = ''
books_details['description'] = ''

# Iterate over each book title and append the genre and description
for i, row in books_details.iloc[6291:].iterrows():
    print(i)
    title = row['Book_Title']
    book_info = get_book_info(title)
    if (i+1) % 100 == 0:
        time.sleep(50)
    if book_info:
        books_details.at[i, 'language'] = book_info['language']
        books_details.at[i, 'page_count'] = book_info['page_count']
        books_details.at[i, 'categories'] = book_info['categories']
        books_details.at[i, 'description'] = book_info['description']
    else:
        books_details.at[i, 'language'] = " "
        books_details.at[i, 'page_count'] = " "
        books_details.at[i, 'categories'] = " "
        books_details.at[i, 'description'] = " "

```
